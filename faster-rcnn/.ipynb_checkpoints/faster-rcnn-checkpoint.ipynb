```python
import cv2
import torch, torchvision
import torch.nn as nn
from torchvision import transforms, models
import matplotlib.pyplot as plt
import os
import json
```

```python
rcnn = torchvision.models.detection.fasterrcnn_resnet50_fpn(weights=models.detection.FasterRCNN_ResNet50_FPN_Weights.DEFAULT)
rcnn
```

```python
for param in rcnn.parameters():
    param.requires_grad = False
```

```python
rcnn.roi_heads.box_predictor.cls_score = nn.Linear(in_features=1024, out_features=7, bias=True)
rcnn.roi_heads.box_predictor.bbox_pred = nn.Linear(in_features=1024, out_features=7*4, bias=True)
rcnn.roi_heads.box_predictor
```
